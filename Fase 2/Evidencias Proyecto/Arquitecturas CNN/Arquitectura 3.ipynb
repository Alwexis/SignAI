Importamos librerías útiles

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
# El import está en desuso, importar cómo lo hice abajo.
# from keras.utils import np_utils
from keras.utils import to_categorical
```

Definimos las rutas del set de datos, set de pruebas y de entrenamientos.

```python
carpeta_base = '/content/ASL_Dataset'
carpeta_entrenamiento = os.path.join(carpeta_base, 'Train')
carpeta_pruebas = os.path.join(carpeta_base, 'Test')
```

Definimos parámetros constantes como el Alto y Ancho de las entradas, también el batch_size de los entrenamientos, el cual, en el caso de esta Topología la definiremos en 64, ya que lo recomendable es utilizar potencias de 2.

```python
alto_entradas = 400
ancho_entradas = 400
batch_size = 16
epochs = 90
```

Creamos el generador de datos de entrenamiento y pruebas

```python
train_datagen = ImageDataGenerator(
    rescale=1./255  # Escalamos los Píxeles de la imágenes
)

test_datagen = ImageDataGenerator(
    rescale=1./255  # Escalamos los Píxeles de las imágenes
)
```

Cargamos los datos de entrenamiento y prueba

```python
train_generator = train_datagen.flow_from_directory(
    carpeta_entrenamiento,
    target_size=(alto_entradas, ancho_entradas),
    batch_size=batch_size,
    class_mode='categorical',  # Nuestor problema es multiclase
    color_mode='rgb'  # Hay que asegurar que las imágenes se carguen en RGB
)

# Hacemos el de Pruebas ahora
test_generator = test_datagen.flow_from_directory(
    carpeta_pruebas,
    target_size=(alto_entradas, ancho_entradas),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,  # No mezclar para evaluar correctamente
    color_mode='rgb'  # Hay que asegurar que las imágenes se carguen en RGB
)
```

Ahora comprobaremos el nombre de las clases de nuestro Set de Datos

MODELO

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization

def _conv_model():
    # Creamos el Modelo
    model = Sequential()

    # Primera capa Conv + BatchNormalization + Pooling
    model.add(Conv2D(16, (3, 3), input_shape=(400, 400, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D())

    # Segunda capa Conv + BatchNormalization + Pooling
    model.add(Conv2D(32, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D())

    # Tercera capa Conv + BatchNormalization + Pooling
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D())

    # Regularización: capa Dropout
    model.add(Dropout(0.5))

    # Aplanamos las características
    model.add(Flatten())

    # Primera capa densa
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))  # Regularización adicional

    # Capa de salida, ajustar el número de neuronas si es necesario
    model.add(Dense(28, activation='softmax'))  # Cambiar 28 si hay más clases

    # Compilamos el Modelo
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

Metricas

```python
model = _conv_model()

model.fit(train_generator, epochs=epochs, validation_data=test_generator, batch_size=batch_size, verbose=2)
```
